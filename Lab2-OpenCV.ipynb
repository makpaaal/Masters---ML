```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
fig= plt.figure(figsize=(15,10))
my_bike = cv2.imread('watch.jpg', cv2.IMREAD_UNCHANGED)
cv2.imshow('my_bike', my_bike)
cv2.imwrite('bike.png', my_bike)
#plt.subplot(122)
plt.imshow(my_bike)
plt.title('Bike')
plt.show()
```

```python
fig= plt.figure(figsize=(15,10))
image_1 = cv2.imread('robo.jpg')
image_2 = cv2.imread('logo.jpg')
result = cv2.add(image_2, image_1)
plt.subplot(131)
plt.imshow(image_1)
plt.title('Robo')
plt.subplot(132)
plt.imshow(image_2)
plt.title('Logo')
plt.subplot(133)
plt.imshow(result)
plt.title('Robo+Logo')
plt.show()
```

```python
fig= plt.figure(figsize=(15,10))
iimage_1 = cv2.imread('robo.jpg')
iimage_2 = cv2.imread('logo.jpg')
result2 = cv2.addWeighted(iimage_1, 0.9, iimage_2, 0.1, 0.2)
plt.subplot(131)
plt.imshow(iimage_1)
plt.title('Robo')
plt.subplot(132)
plt.imshow(iimage_2)
plt.title('Logo')
plt.subplot(133)
plt.imshow(result2)
plt.title('Robo+Logo')
plt.show()
```

```python
original_image = cv2.imread('watch.jpg')
average_image = cv2.blur(original_image,(3,3))
gaussian_image = cv2.GaussianBlur(original_image,(3,3),0)
median_image = cv2.medianBlur(original_image,3)

fig= plt.figure(figsize=(15,10))

plt.subplot(221)
plt.imshow(original_image)
plt.title('Original')
plt.subplot(222)
plt.imshow(average_image)
plt.title('Blur')
plt.subplot(223)
plt.imshow(gaussian_image)
plt.title('GaussianBlur')
plt.subplot(224)
plt.imshow(median_image)
plt.title('MedianBlur')
plt.show()
```

```python
image = cv2.imread('watch.jpg', cv2.IMREAD_UNCHANGED)
result_1 = cv2.resize(image, None, fx=2, fy=10, interpolation=cv2.INTER_CUBIC)
result_2 = cv2.resize(image, None, fx=5, fy=2, interpolation=cv2.INTER_AREA)

plt.subplot(121)
plt.imshow(result_1)
plt.title('Scale up/expand')
plt.subplot(122)
plt.imshow(result_2)
plt.title('Scale down/shrink')
plt.show()
```

```python
img = cv2.imread('watch.jpg', cv2.IMREAD_UNCHANGED)
width = 440
height = img.shape[0] # keep original height
dim = (width, height)
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

plt.imshow(resized)
plt.show()
```

```python
image = cv2.imread('watch.jpg',0)
r, c = image.shape
matrix = cv2.getRotationMatrix2D((c/2,r/2), 180, 1)
result = cv2.warpAffine(image,matrix,(c,r))

plt.imshow(result)
plt.show()
```
