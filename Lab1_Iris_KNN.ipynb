```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
#import mglearn
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print("Кeys iris_dataset: \n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset['DESCR'][:1160] + "\n...")
```

```python
print("Classes: {}".format(iris_dataset['target_names']))
print("Features: \n{}".format(iris_dataset['feature_names']))
```

```python
print("First 5 rows:\n{}".format(iris_dataset['data'][:5]))
print("Answers:\n{}".format(iris_dataset['target'][:5]))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print("X_train: {}".format(X_train.shape))
print("y_train: {}".format(y_train.shape))
print("X_test: {}".format(X_test.shape))
print("y_test: {}".format(y_test.shape))
```

```python
# создаем dataframe из данных в массиве X_train
# маркируем столбцы, используя строки в iris_dataset.feature_names
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
# создаем матрицу рассеяния из dataframe, цвет точек задаем с помощью y_train
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', alpha=.8)
plt.show()
```

```python
knn5 = KNeighborsClassifier(n_neighbors=5)
knn3 = KNeighborsClassifier(n_neighbors=3)
knn7 = KNeighborsClassifier(n_neighbors=7)
knn5.fit(X_train, y_train)
```

```python
knn3.fit(X_train, y_train)
```

```python
knn7.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
```

```python
prediction = knn5.predict(X_new)
print("Прогноз: {}".format(prediction))
print("Спрогнозированная метка: {}".format(
 iris_dataset['target_names'][prediction]))
```

```python
prediction3 = knn3.predict(X_new)
print("Прогноз: {}".format(prediction3))
print("Спрогнозированная метка: {}".format(
 iris_dataset['target_names'][prediction3]))
```

```python
prediction7 = knn7.predict(X_new)
print("Прогноз: {}".format(prediction7))
print("Спрогнозированная метка: {}".format(
 iris_dataset['target_names'][prediction7]))
```

```python
y_pred = knn5.predict(X_test)
print("Прогнозы для тестового набора:\n {}".format(y_pred))
```

```python
y_pred3 = knn3.predict(X_test)
print("Прогнозы для тестового набора:\n {}".format(y_pred3))
```

```python
y_pred7 = knn7.predict(X_test)
print("Прогнозы для тестового набора:\n {}".format(y_pred7))
```

```python
print("Accuracy for k=5 using np.mean: {:.2f}".format(np.mean(y_pred == y_test)))
print("Accuracy for k=3 using np.mean: {:.2f}".format(np.mean(y_pred3 == y_test)))
print("Accuracy for k=7 using np.mean: {:.2f}".format(np.mean(y_pred7 == y_test)))

print("Accuracy for k=5 using knn.score: {:.2f}".format(knn5.score(X_test, y_test)))
print("Accuracy for k=3 using knn.score: {:.2f}".format(knn3.score(X_test, y_test)))
print("Accuracy for k=7 using knn.score: {:.2f}".format(knn7.score(X_test, y_test)))
```
